```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.applications.mobilenet import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
import seaborn as sns

print("=" * 50)
print("FAST DEEP LEARNING MODEL + AUGMENTATION (MobileNetV1)")
print("=" * 50)

print("\n[1] Loading CIFAR-10...")

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

x_train = x_train[:3000]
y_train = y_train[:3000]

x_test = x_test[:500]
y_test = y_test[:500]

print("Training:", x_train.shape)
print("Testing :", x_test.shape)

print("\n[2] Resizing images...")

IMG_SIZE = 96

x_train_resized = np.array([
    cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    for img in x_train
])

x_test_resized = np.array([
    cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    for img in x_test
])

print("\n[3] Preprocessing...")

x_train_resized = preprocess_input(
    x_train_resized.astype(np.float32)
)

x_test_resized = preprocess_input(
    x_test_resized.astype(np.float32)
)

y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("\n[4] Building MobileNetV1 model...")

base_model = MobileNet(
    weights='imagenet',
    include_top=False,
    input_shape=(96, 96, 3)
)

base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n[5] Data Augmentation Setup...")

datagen = ImageDataGenerator(
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.2
)

datagen.fit(x_train_resized)

print("\n[6] Training started...")

start = time.time()

history = model.fit(
    datagen.flow(x_train_resized, y_train_cat, batch_size=32),
    validation_data=(x_test_resized, y_test_cat),
    epochs=3,
    verbose=1
)

training_time = time.time() - start

print(f"\nTraining Time: {training_time:.2f} seconds")

print("\n[7] Evaluating...")

predictions = model.predict(x_test_resized, verbose=0)

pred_classes = np.argmax(predictions, axis=1)

accuracy = accuracy_score(y_test, pred_classes)
precision = precision_score(y_test, pred_classes, average='weighted')
recall = recall_score(y_test, pred_classes, average='weighted')
f1 = f1_score(y_test, pred_classes, average='weighted')

print("\n" + "=" * 50)
print("RESULTS")
print("=" * 50)

print(f"Accuracy : {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall   : {recall*100:.2f}%")
print(f"F1-Score : {f1*100:.2f}%")

print("\nClassification Report:\n")

print(classification_report(
    y_test,
    pred_classes,
    target_names=[
        'airplane','automobile','bird','cat','deer',
        'dog','frog','horse','ship','truck'
    ]
))

print("\nConfusion Matrix:\n")

cm = confusion_matrix(y_test, pred_classes)
print(cm)

# =====================================================
# =====================================================

print("\n[8] Saving Confusion Matrix image to your computer...")

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(12, 10))

sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Reds',
            xticklabels=class_names,
            yticklabels=class_names,
            square=True,
            linewidths=0.5,
            linecolor='black')

plt.title('Confusion Matrix - End-to-End Deep Learning (MobileNetV1)', fontsize=16, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()

plt.savefig('confusion_matrix_ende2e.png', dpi=300, bbox_inches='tight')
print("\n✅ Image saved as: confusion_matrix_ende2e.png")

print("\n✅ Model training completed successfully!")
```

```python

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import Model

from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix
)

import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
import seaborn as sns

print("=" * 50)
print("FEATURE EXTRACTION + SVM")
print("=" * 50)

# 1. Load CIFAR-10
print("\n[1] Loading CIFAR-10...")
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

x_train = x_train[:3000]
y_train = y_train[:3000]
x_test = x_test[:500]
y_test = y_test[:500]

print(f"Training: {x_train.shape}")
print(f"Testing: {x_test.shape}")

# 2. Resize images
print("\n[2] Resizing images to 96x96...")
IMG_SIZE = 96
x_train_resized = np.array([cv2.resize(img, (IMG_SIZE, IMG_SIZE)) for img in x_train])
x_test_resized = np.array([cv2.resize(img, (IMG_SIZE, IMG_SIZE)) for img in x_test])

# 3. Preprocessing
print("\n[3] Preprocessing...")
x_train_resized = preprocess_input(x_train_resized.astype(np.float32))
x_test_resized = preprocess_input(x_test_resized.astype(np.float32))

# 4. Load MobileNetV2 as feature extractor
print("\n[4] Loading MobileNetV2...")
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(96, 96, 3))
feature_extractor = Model(inputs=base_model.input, outputs=base_model.output)

# 5. Extract features
print("\n[5] Extracting features...")
start = time.time()
train_features = feature_extractor.predict(x_train_resized, verbose=0)
test_features = feature_extractor.predict(x_test_resized, verbose=0)

train_features = train_features.reshape(train_features.shape[0], -1)
test_features = test_features.reshape(test_features.shape[0], -1)

print(f"Feature Extraction Time: {time.time() - start:.2f} seconds")

# 6. Train SVM
print("\n[6] Training SVM...")
start = time.time()
svm = SVC(kernel='linear')
svm.fit(train_features, y_train.ravel())
print(f"SVM Training Time: {time.time() - start:.2f} seconds")

# 7. Evaluate
print("\n[7] Evaluating...")
predictions = svm.predict(test_features)

print("\n" + "=" * 50)
print("RESULTS")
print("=" * 50)
print(f"Accuracy : {accuracy_score(y_test, predictions)*100:.2f}%")
print(f"Precision: {precision_score(y_test, predictions, average='weighted')*100:.2f}%")
print(f"Recall   : {recall_score(y_test, predictions, average='weighted')*100:.2f}%")
print(f"F1-Score : {f1_score(y_test, predictions, average='weighted')*100:.2f}%")

print("\nClassification Report:")
print(classification_report(y_test, predictions, target_names=[
    'airplane', 'automobile', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]))

# 8. Save Confusion Matrix as image
print("\n[8] Saving Confusion Matrix...")
cm = confusion_matrix(y_test, predictions)
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names,
            square=True, linewidths=0.5, linecolor='black')
plt.title('Confusion Matrix - Feature Extraction + SVM', fontsize=16)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('confusion_matrix_approach1.png', dpi=300, bbox_inches='tight')
print("✅ Image saved as: confusion_matrix_approach1.png")

print("\n✅ Feature Extraction + SVM completed successfully!")
```
