```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import ResNet50, MobileNet

from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
import time

os.makedirs("results", exist_ok=True)

IMG_SIZE = (224, 224)
INPUT_SHAPE = (224, 224, 3)

EPOCHS = 5
BATCH_SIZE = 32
LR = 0.001


# ================= DATA =================
def load_cifar10():
    (x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

    x_train = tf.image.resize(x_train[:1000].astype("float32")/255, IMG_SIZE).numpy()
    x_test = tf.image.resize(x_test[:200].astype("float32")/255, IMG_SIZE).numpy()

    return x_train, y_train[:1000].flatten(), x_test, y_test[:200].flatten()


# ================= METRICS =================
def metrics(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average='weighted'),
        "Recall": recall_score(y_true, y_pred, average='weighted'),
        "F1": f1_score(y_true, y_pred, average='weighted')
    }


# ================= CONFUSION MATRIX =================
def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues")

    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")

    plt.savefig(f"results/{title}.png")
    plt.show()


# ================= METRICS PLOT (BLUE & WHITE) =================
def plot_metrics(r_svm, r_dl, m_svm, m_dl, title="Metrics_Comparison"):

    labels = ["Accuracy", "Precision", "Recall", "F1"]
    x = np.arange(len(labels))

    def vals(d):
        return [d["Accuracy"], d["Precision"], d["Recall"], d["F1"]]

    plt.figure(figsize=(10,5))

    plt.bar(x - 0.3, vals(r_svm), 0.2, label="ResNet SVM", color="#1f77b4", edgecolor="white")
    plt.bar(x - 0.1, vals(r_dl), 0.2, label="ResNet DL", color="#aec7e8", edgecolor="white")
    plt.bar(x + 0.1, vals(m_svm), 0.2, label="MobileNet SVM", color="#1f77b4", alpha=0.7, edgecolor="white")
    plt.bar(x + 0.3, vals(m_dl), 0.2, label="MobileNet DL", color="#aec7e8", alpha=0.9, edgecolor="white")

    plt.xticks(x, labels)
    plt.title(title)

    plt.gca().set_facecolor("white")
    plt.grid(axis='y', linestyle='--', alpha=0.3)

    plt.legend()

    plt.savefig(f"results/{title}.png", bbox_inches='tight')
    plt.show()


# ================= FEATURE + SVM =================
def feature_svm(model, x_train, y_train, x_test, y_test):
    model.trainable = False

    f_train = model.predict(x_train, verbose=0).reshape(len(x_train), -1)
    f_test = model.predict(x_test, verbose=0).reshape(len(x_test), -1)

    scaler = StandardScaler()
    f_train = scaler.fit_transform(f_train)
    f_test = scaler.transform(f_test)

    svm = SVC(kernel='linear')
    svm.fit(f_train, y_train)

    pred = svm.predict(f_test)

    return metrics(y_test, pred), pred


# ================= END-TO-END =================
def end_to_end(base_model, num_classes, x_train, y_train, x_test, y_test):

    base_model.trainable = False

    model = keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(LR),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    model.fit(
        x_train, y_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=0
    )

    pred = np.argmax(model.predict(x_test), axis=1)

    return metrics(y_test, pred), pred


# ================= RUN =================
def main():

    x_train, y_train, x_test, y_test = load_cifar10()

    resnet = ResNet50(weights="imagenet", include_top=False, input_shape=INPUT_SHAPE)
    mobile = MobileNet(weights="imagenet", include_top=False, input_shape=INPUT_SHAPE)

    # ===== RESNET =====
    r_svm, r_pred_svm = feature_svm(resnet, x_train, y_train, x_test, y_test)
    r_dl, r_pred_dl = end_to_end(resnet, 10, x_train, y_train, x_test, y_test)

    # ===== MOBILE =====
    m_svm, m_pred_svm = feature_svm(mobile, x_train, y_train, x_test, y_test)
    m_dl, m_pred_dl = end_to_end(mobile, 10, x_train, y_train, x_test, y_test)

    # ===== CONFUSION MATRICES =====
    plot_cm(y_test, r_pred_svm, "CIFAR_ResNet_SVM")
    plot_cm(y_test, r_pred_dl, "CIFAR_ResNet_DL")
    plot_cm(y_test, m_pred_svm, "CIFAR_MobileNet_SVM")
    plot_cm(y_test, m_pred_dl, "CIFAR_MobileNet_DL")

    # ===== METRICS COMPARISON =====
    plot_metrics(r_svm, r_dl, m_svm, m_dl)

    print("DONE CIFAR-10 ✔")


if __name__ == "__main__":
    main()
```
